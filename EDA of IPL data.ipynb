```python
## Check if column 'umpire3' is useful. If not drop it and save the dataframe to a file.
df_matches['umpire3'].nunique()
df2=df_matches.drop(['umpire3'],axis=1)
df2.to_csv('matches2.csv',index=False)
```

```python
## Print Total Number of Mataches played, Number of Seasons, 
print('Total no of matches:',df_matches['id'].nunique())#or df_matches.shape[0]
print('No of Seasons:',df_matches['season'].nunique())
```

```python
# Players who have won the most "Player of the Match" 
## And Team which has the highest number of match wins
print('won most of the times as "player of the match":',df_matches['player_of_match'].value_counts().idxmax())
print('most player of the match:',df_matches['player_of_match'].value_counts()[:10] )
print('highest number of match wins:',df_matches['winner'].value_counts().idxmax())
```

```python
#Find Teams which have won matches by a huge margin 
huge_margin=df_matches[(df_matches['win_by_runs']>=100)|(df_matches['win_by_wickets']>=8)]
huge_margin['winner'].value_counts()
```

```python
# Find Count of Matches Played at each city 
df_matches['city'].value_counts()
```

```python
#Find Min and Max win_by_wickets at each city
min_wickets= df_matches[df_matches['win_by_wickets']<=3]
max_wickets= df_matches[df_matches['win_by_wickets']>=8]
print('min win by wickets at each city')
print(min_wickets['city'].value_counts())
print('Max win by wickets at each city')
print(max_wickets['city'].value_counts())
```
